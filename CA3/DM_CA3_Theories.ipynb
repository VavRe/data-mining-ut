```python
points = [(0.18,0.76),(0.02,0.27),(0.55,0.52),(0.88,0.53),(0.38,0.77),(0.35,0.05)]
points
```

```python
import numpy as np
import pandas as pd
euclidean_dist = {point: [] for point in points}
manhattan_dist = {point: [] for point in points}
for point in points:
  for second_point in points:
    dx = np.abs(point[0] - second_point[0])
    dy = np.abs(point[1] - second_point[1])
    euclidean_dist[point].append(np.round(np.sqrt(dx**2+dy**2),4)) 
    manhattan_dist[point].append(dx + dy) 
  
```

```python
pd.DataFrame(euclidean_dist)
```
